```python
# Let's begin by extracting the text from the PDF and focusing on the relevant parts.
# Importing necessary libraries to handle the PDF and extract text
import PyPDF2

# Path to the uploaded PDF file
pdf_path = '/content/T.E.(2019 PAT.)(INFORMATION TECHNOLOGY).pdf'

# Function to extract text from the PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

# Extracting the text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Display first 2000 characters of the extracted text to understand the structure
text=pdf_text[:]
text

text_file_path = '/content/extracted_pdf_text.txt'
with open(text_file_path, 'w') as text_file:
    text_file.write(text)

text_file_path
```

```python
import re
import csv
import json  # To format dictionaries in CSV

# Function to extract Course Name, ISE, and ESE marks by Seat Number and Semester
def extract_seat_number_wise_results(text):
    # Splitting the text into lines for easier processing
    lines = text.split("\n")

    extracted_data = {}
    seat_number = None
    current_semester = None

    # Iterate over each line to extract the required information
    for line in lines:
        # Check for seat number (e.g., SEAT NO.: T190318501)
        seat_match = re.search(r'SEAT NO\.\s*:\s*(T\d+)', line)
        if seat_match:
            seat_number = seat_match.group(1)
            # Initialize a new entry for the seat number if not already present
            if seat_number not in extracted_data:
                extracted_data[seat_number] = {'Semester': {}}
            continue

        # Check for semester (e.g., SEM.:1 or SEM.:2)
        semester_match = re.search(r'SEM\.\s*:\s*(\d)', line)
        if semester_match:
            current_semester = f"SEM.:{semester_match.group(1)}"
            # Initialize a dictionary for the current semester if not already present
            if current_semester not in extracted_data[seat_number]['Semester']:
                extracted_data[seat_number]['Semester'][current_semester] = {}
            continue

        # Check for Course Name and Marks (ISE and ESE)
        course_match = re.search(r'^\s*(\d{6}[A-Z]?)\s+([A-Z\s\(\)-]+?)\s+(\d{3}/030)\s+(\d{3}/070)\s+', line)
        if course_match and seat_number and current_semester:
            course_code, course_name, ise_marks, ese_marks = course_match.groups()
            # Add course details under the current semester
            extracted_data[seat_number]['Semester'][current_semester][course_name.strip()] = {
                'ISE': int(ise_marks[:3]),
                'ESE': int(ese_marks[:3])
            }

    print(extracted_data)
    return extracted_data

# Function to save extracted data into a CSV file
def save_to_csv(data, output_file):
    # Open CSV file to write data
    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)

        # Write headers
        writer.writerow(['Seat Number', 'Semester', 'Subjects'])

        # Write data
        for seat_number, seat_data in data.items():
            for semester, semester_data in seat_data['Semester'].items():
                # Convert the subjects and their ISE/ESE marks into a string representation
                subjects = json.dumps(semester_data)
                writer.writerow([seat_number, semester, subjects])

# Sample extracted text from the PDF (you should replace this with the actual content of the PDF text)
# pdf_text = """
# SEAT NO.: T190318501
# SEM.:1
#    314445B ADVANCED DATABASE MGMT SYS       025/030  070/070
#    314444 HUMAN COMPUTER INTERACTION       029/030  062/070
# SEM.:2
#    314460 LABORATORY PRACTICE-II           ---      ---      ---    046/050  022/025
#    314454C CLOUD COMPUTING                030/030  052/070  082/100
# SEAT NO.: T190318502
# SEM.:1
#    314445B ADVANCED DATABASE MGMT SYS       027/030  070/070
#    314444 HUMAN COMPUTER INTERACTION       030/030  070/070
#    314443 MACHINE LEARNING                 029/030  064/070
# SEM.:2
#    314460 LABORATORY PRACTICE-II           ---      ---      ---    034/050  020/025
#    314454B CYBER SECURITY                 030/030  046/070  076/100
# """

pdf_text = text

# Extracting the data from the PDF text
seat_number_wise_results = extract_seat_number_wise_results(pdf_text)

# Saving the extracted data to a CSV file
output_csv_file = 'seat_number_wise_results.csv'
save_to_csv(seat_number_wise_results, output_csv_file)

print(f"Data saved to {output_csv_file}")
```
